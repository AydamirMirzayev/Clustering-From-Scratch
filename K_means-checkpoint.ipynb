```python
# K means implementation with numpy 
import numpy as np 
import pandas as pd 
import random as rnd
import math
import matplotlib.pyplot as plt 
 
"""
Code to hold and test from scratch implementation of K-means algolithm of a 2D dataset

Authors
-------
Aydamir Mirzayev: https://github.com/AydamirMirzayev || https://www.linkedin.com/in/aydamir-mirzayev-97b297133/

Affiliation
-----------
Part of Bilkent University database

Methods
-------
standardize(data)
    Standizes the input data across every dimension: (x-mu)/std

K_means(k,X,epoc,treshold)
    Iteratively finds the mean of k clusters using expectation maximization
    Method assumes input dataset of 2 feature dimentions. Plottable on a 2D scatterplot.
    
visaulize_k_means_result(data, MU, title)
    Visualizes clusters reuslting from the K-means algorithm
"""    
```

```python
def standardize(data):
    
    """ Standizes the input data across every dimension: (x-mu)/std
        
    Parameters:
    data (2D int array): 2 dimentional input samples 
    
    Returns:
    new_data (2D int array): standardize version of the data 
   """
    
    L, dim = data.shape # L: number of instances and dim: number of data dimensions 
    new_data = np.zeros( data.shape) # placeholder for new_data 
    
    # standardize each dimension separately
    for i in range( dim):
        mean = np.mean(data[:,i])
        new_data[:,i] = (data[:,i] - mean)/ np.std( data[:,i])
    
    return new_data
```

```python
def K_means(k,X,epoc,treshold):
    
    """ Iteratively finds the mean of k clusters using expectation maximization
        Method assumes input dataset of 2 feature dimentions. Plottable on a 2D scatterplot.
        
    Parameters:
    k          (int): number of clusters to be used
    X (2D int array): 2 dimentional input samples 
    epoc       (int): maximum number of iterations in the EM step
    treshold (float): stopping treshold (if the absolute change of mean along all dimensions is less than treshold)
    
    Returns:
    MU (2D int array): k to 2 array of final cluster means. 
   """
    
    # param
    length = X.shape[0] #number of data instances
    MU = np.zeros((2,k)) #mean array
    distance = np.zeros((length,k)) # all cluster-point distances
    to_break = True #stopping flag initially set to positive 
    
    # initialize the means randomly in the max-min range across both dimensions 
    min_x1 = np.min(X[:,0]) 
    max_x1 = np.max(X[:,0])
    min_x2 = np.min(X[:,1])
    max_x2 = np.max(X[:,1])
    for i in range(k):
        MU[0,i] = rnd.random()*(max_x1 - min_x1) + min_x1
        MU[1,i] = rnd.random()*(max_x2 - min_x2) + min_x2
        
    # EM step 
    for e in range(epoc):
        # calculate the cluster mean to point distances for all cluster-point pairs.
        for n in range(length):
            for i in range(k):
                # Euclidean distance is used
                dist = math.sqrt( (X[n,0]-MU[0,i])**2 + (X[n,1]-MU[1,i])**2)
                distance[n,i] = dist
        
        # find the neares cluster mean
        cluster = np.argmin(distance, 1).reshape(length,1)
        
        # calculate new cluster means
        for i in range(k):
            
            indices = np.array([j for j in range(0,length) if cluster[j]== i])
            if indices.shape[0] != 0:
                muk = np.mean(X[indices], 0).reshape(2,1)
                
                if( abs(MU[0,i] - muk[0,0]) > treshold or abs(MU[1,i] - muk[1,0]) > treshold):
                    to_break = False
                    
                MU[0,i] = muk[0,0]
                MU[1,i] = muk[1,0]
        
        # check whether to stop
        if (to_break):
            print('Done at iteration no:', e)
            break
        
        to_break = True
    
    return MU
```

```python
def visaulize_k_means_result(data, MU, title):
    
    
    """ Visualizes clusters reuslting from the K-means algorithm
        
    Parameters:
    data (2D int array): 2 dimentional input samples 
    MU   (2D int array): k to 2 array of final cluster means. 
    title      (string): title of the plot
    
   """
    
    no_of_clusters = MU.shape[1] #number of clusters
    length = data.shape[0] #number of data samples 
    colors = [ 'goldenrod', 'green', 'purple' , 'gray','orange', 'yellow', 'royalblue', 'cyan', 'goldenrod']
    
    
    
    # calculate distance of each cluster meand - point pair
    distance = np.zeros((length,no_of_clusters)) # placeholders for distance of each cluster meand - point pair
    for n in range(length):
        for i in range(no_of_clusters):
            dist = math.sqrt( (data[n,0]-MU[0,i])**2 + (data[n,1]-MU[1,i])**2)
            distance[n,i] = dist
    
    # array indication to which cluster each point belongs to
    point_clusters_assignment = np.argmin(distance, 1).reshape(length,1)
    
    
    #plot points in each cluster with different colors
    plt.figure() 
    for i, u in enumerate(np.unique( point_clusters_assignment)):
        indexes = np.argwhere( point_clusters_assignment ==u)[:,0]
        plt.plot(data[indexes,0],data[indexes,1], '*', color= colors[i])
    
    #plot cluster means
    for e in range(no_of_clusters):   
        plt.plot( MU[0,e], MU[1,e], 'o', color='black')
        
    # adjust the style (non-functional, for better visualization)
    plt.title(title, fontsize=15)
    plt.xticks(fontsize=21)
    plt.yticks(fontsize=21)
    
    
    #arrange x tick spacing (non-functional, for better visualization)
    xtics = np.zeros(5)
    xmax = max(data[:,0])
    xmin = min(data[:,0])
    step = (xmax - xmin)/4
    
    curr = xmin
    for i in range(5):
        xtics[i] = round(curr, 2)
        curr = curr + step
        
    #arrange  y tick spacing (non-functional, for better visualization)
    ytics = np.zeros(5)
    ymax = max(data[:,1])
    ymin = min(data[:,1])
    step = (ymax - ymin)/4
    
    curr = ymin
    for i in range(5):
        ytics[i] = round(curr, 2)
        curr = curr + step    
    
    # plot the tics (non-functional, for better visualization)
    plt.xticks( xtics )
    plt.yticks( ytics )
    
    return point_clusters_assignment
```

```python
# Read and standardize the data
data1 = np.array( pd.read_csv("data1.txt", header = None, delim_whitespace=True, skipinitialspace=True), dtype=np.double)
data2 = np.array( pd.read_csv("data2.txt", header = None, delim_whitespace=True, skipinitialspace=True), dtype=np.double)

data1 = standardize(data1)
data2 = standardize(data2)
```

```python
# Test the methods 
MU = K_means( 2, data1, 100, 0.0001) #test first dataset with 3 clusters 
_ = visaulize_k_means_result(data1, MU, '1st Data Set K-means Clustering')

MU = K_means( 3, data2, 100, 0.0001) #test third dataset with 7 clusters 
_ = visaulize_k_means_result(data2, MU, '2nd Data Set K-means Clustering')
```
